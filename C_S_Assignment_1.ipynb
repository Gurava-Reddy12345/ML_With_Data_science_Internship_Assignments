```python
import pandas as pd
import numpy as np
df=pd.read_csv("Asg1_Data.csv")
df
```

```python
df['Name'] = df['Name'].astype(str).str.strip().str.title()
df['Name'] = df['Name'].replace("Nan", "Unnamed")

# 2. Handle missing Names
df['Name'] = df['Name'].replace('', 'Unnamed')

# 3. Convert Region to uppercase and handle missing Region
df['Region'] = df['Region'].str.upper()
df['Region'] = df['Region'].fillna('Unknown')
print("Region:-")
print(df["Region"])
# 4. Replace missing Marks with average
average_marks = df['Marks'].mean()
df['Marks'] = df['Marks'].fillna(average_marks)

# 5. Convert Date column to dd-mm-yyyy
df['Date'] = pd.to_datetime(df['Date'], errors='coerce').dt.strftime('%d-%m-%Y')

# 6. Remove duplicate rows
df = df.drop_duplicates()

# 7. Create Result column: Pass if Marks >= 70
df['Result'] = df['Marks'].apply(lambda x: 'Pass' if x >= 70 else 'Fail')

# 8. Summary: Average marks by Region
avg_marks_by_region = df.groupby('Region')['Marks'].mean().reset_index()

# 9. Summary: Number of students passed in each Subject
pass_count_by_subject = df[df['Result'] == 'Pass'].groupby('Subject').size().reset_index(name='no.of passed students')
```
